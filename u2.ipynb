# Plano de fase

```python
t = sp.symbols("t")
f=10; # frecuencia de la seal en Hz 
w=2*sp.pi*f; # Frecuencia de la seal en rad/s
Fs=32000; # Frecuencia de muestreo en Hz
tf=1.5; # Duracin de la simulacin

x = 4*sp.sin(w*t)*sp.exp(-5*t) # vector de posicin
dxdt = 4*sp.exp(-5*t)*(-5*sp.sin(w*t) + w*sp.cos(w*t)) # derivada de la posicin = velocidad

sp.plot_parametric(x, dxdt, (t, 0, tf))

```
